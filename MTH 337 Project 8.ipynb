```python
import requests
import re
```

```python
# Here, I pull the directory of UB's School of Pharmacy
directory = requests.get('https://pharmacy.buffalo.edu/faculty-staff.html').text
print(directory)
```

```python
urlPattern = r'href="(https://pharmacy\.buffalo\.edu/content/pharmacy/faculty-staff/faculty-profile\.html\?ubit=\w+)'

# This returns all the urls for each person in the directory
```

```python
URLS = re.findall(urlPattern, directory)
print(URLS)
```

```python
# Here, I split the html code into multiple lines
test = requests.get('https://pharmacy.buffalo.edu/content/pharmacy/faculty-staff/faculty-profile.html?ubit=kadams4').text
test = test.replace('>', '>\n')
print(test)
```

```python
# This will cut down the amount of text I need to sift through
startSplit = '''<span id="lblFacultyName">'''
endSplit = '''<div class="facultyprofilemapicon">'''
```

```python
kimberly = (test.split(startSplit)[1].split(endSplit)[0])
print(kimberly)
phoneSplit = r'Phone:'
#print(phoneSplit)
officeSplit = test.split(startSplit)[1].split(phoneSplit)[0]
#print(officeSplit)
```

```python
print(officeSplit)
```

```python
namePattern = r'\s\<span id\=\"lblFacultyNameMobile\"\>\n(.*)</span>'  
re.findall(namePattern, kimberly)
```

```python
officePattern = r'\<span id\=\"lblAcademicOfficeAddressMobile\"\>\n(?:(.*)\<br/\>\n)*' # this pulls the offices
re.findall(officePattern, officeSplit)
```

```python
phonePattern = r'\<br\/\>\n(Phone:.*)<br/>' # this pulls the phone numbers
re.findall(phonePattern, kimberly)
```

```python
emailPattern = r'"mailto:(\w+\@\w+\.\w+)' # this pulls the emails
re.findall(emailPattern, kimberly)
```

```python
for url in URLS:                                        # iterates over the list of urls
    text = session.get(url).text
    text = text.replace('>','>\n')                      # splits the html to make it mroe readable
    names = re.findall(namePattern, text)               # grabs all the names
    if len(names) > 0:
        print(names)
        textSplit = text.split(startSplit)[1].split(endSplit)[0] # removes superfluous html
        offices = re.findall(officePattern, officeSplit)           # grabs all the offices
        phones = re.findall(phonePattern, textSplit)             # grabs all the phone numbers
        emails = re.findall(emailPattern, textSplit)             # grabs all the emails
        #page = page.split(startSplit)[1].split(endSplit)[0] # removes superfluous html
        if len(offices) > 0:
            print(offices)
        else:
            print('No office listed', url)
        if len(phones) > 0:
            print(phones)
        else:
            print('No phone number listed', url)
        if len(emails) > 0:
            print(emails)
        else:
            print('No email listed', url)
    else:
        print('no name found', url)
    #page = page.split(startSplit)[1].split(endSplit)[0] # removes superfluous html
    #names = re.findall(namePattern, page)               # grabs all the names
    #offices = re.findall(officePattern, page)           # grabs all the offices
    #phones = re.findall(phonePattern, page)             # grabs all the phone numbers
    #emails = re.findall(emailPattern, page)             # grabs all the emails
    #if len(offices) > 0:
     #   print(offices)
    #else:
        #print('No office listed', url)
    #if len(phones) > 0:
        #print(phones)
    #else:
        #print('No phone number listed', url)
    #if len(emails) > 0:
        #print(emails)
    #else:
        #print('No email listed', url)
                                               
    print('-'*100)
```
